```python
from datetime import datetime

def create_title(title):
    upper_title = title.upper()
    padding = ' ' * 4

    formatted_title = f"\n{padding}{upper_title}{padding}\n"
    return formatted_title

title = "Blood Pressure Tracker!"
print(create_title(title))

about = """
    About Blood Pressure:
    
    Blood pressure is the force of blood pushing against the walls of the arteries as your heart pumps blood. It includes two 
    measurement; "Systolic" is your blood pressure when your heart beats while pumping bloods and "Diastolic" is your blood pre-
    sure when the heart us at rest between beats. 

    The goal of the Blood Pressure Tracker is to monitor your health trends, manage your hypertension, and the ability to provide this data
    to your healthcare provider
                """
print(about)

def save_data(name, systolic, diastolic, notes):
    date_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    with open("blood_pressure_data.txt", "a") as file:
        file.write(f"Date/Time: {date_time}\n")
        file.write(f"Name: {name}\n")
        file.write(f"Systolic: {systolic}\n")
        file.write(f"Diastolic: {diastolic}\n")
        file.write(f"Notes: {notes}\n")
        file.write("-" * 30 + "\n")
        
def view_data():
    try:
        with open("blood_pressure_data.txt", "r") as file:
            print("\nPrevious Blood Pressure Readings:")
            print(file.read())
    except FileNotFoundError:
        print("\nNo previous blood pressure readings found.")    
        
def user_input():
    sys = input("Enter systolic reading: ")
    dia = input("Enter diastolic reading: ")
    note = input("Enter any notes thats relevant to your blood pressure (optional) ")

    if not sys or not dia:
        print("Please enter both the systolic and diastolic for reading")
        return
    print("Blood pressure reading saved succesfully")

def main():
    name = input("Enter your name: ")
    print(f"Welcome {name} to the Blood Pressure Tracker!")
    while True:
        print("\nMenu:")
        print("1. Enter Blood Pressure Reading")
        print("2. View prior Blood Pressure Reading")
        print("3. Exit")
        choice = input("Enter your choice (1, 2, or 3):    ")

        if choice == "1":
           user_input()
        elif choice == "2":
             view_data()
        elif choice == "3":
             print("Exiting Blood Pressure Tracker")
             break
        else:
            print("Invalid choice. Please enter 1, 2, or 3")

if __name__ == "__main__":
    main()

        
```
