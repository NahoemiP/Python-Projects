```python
def create_title(title):
    upper_title = title.upper()
    padding = ' ' * 4

    formatted_title = f"\n{padding}{upper_title}{padding}\n"
    return formatted_title

title = "Health Smart Quiz: Test Your Wellness Knowledge!"
print(create_title(title))

instructions = """
                Read the following instructions:
                
                1. Read each question carefully
                2. Choose the correct answer from the give options (A, B, C, or D)
                """
print(instructions)

questions = [
    {
        "prompt": "1. What is the recommended amount of daily physical activity for adults?",
        "options": ["A. 30 minutes", "B. 60 minutes", "C. 45 minutes", "D. 90 minutes"],
        "answer": "A"
    },
    {
        "prompt": "2. Which vitamin is essential for the absorption of calcium in the body?",
        "options": ["A. Vitamin A", "B. Vitamin C", "C. Vitamin D", "D. Vitamin E"],
        "answer": "C"
    },
    {
        "prompt": "3. What is the normal range for an adultâ€™s resting heart rate?",
        "options": ["A. 40-60 beats per minute", "B. 60-100 beats per minute", "C. 80-120 beats per minute", "D. 100-140 beats per minute"],
        "answer": "B"
    },
    {
        "prompt": "4. Which nutrient is most important for muscle repair and growth?",
        "options": ["A. Carbohydrates", "B. Fats", "C. Proteins", "D. Vitamins"],
        "answer": "C"
    },   
    {
        "prompt": "5. Which disease is characterized by high levels of sugar in the blood?",
        "options": ["A. Hypertension", "B. Diabetes", "C. Hyperlipidemia", "D. Osteoporosis"],
        "answer": "B"
    },
    {
        "prompt": "6. How many hours of sleep are recommended for adults each night?",
        "options": ["A. 5-6 hours", "B. 6-7 hours", "C. 7-9 hours", "D. 9-10 hours"],
        "answer": "C"
    },
    {
        "prompt": "7. What is the primary function of red blood cells?",
        "options": ["A. Fighting infections", "B. Carrying oxygen", "C. Clotting blood", "D. Producing hormones"],
        "answer": "B"
    },
    {
        "prompt": "8. Which type of fat is considered the healthiest?",
        "options": ["A. Saturated fats", "B. Trans fats", "C. Monounsaturated fats", "D. Polyunsaturated fats"],
        "answer": "C"
    },
    {
        "prompt": "9. What is the main organ responsible for detoxifying the body?",
        "options": ["A. Kidneys", "B. Liver", "C. Lungs", "D. Heart"],
        "answer": "B"
    },
    {
        "prompt": "10. Which mineral is important for maintaining strong bones and teeth?",
        "options": ["A. Iron", "B. Magnesium", "C. Calcium", "D. Potassium"],
        "answer": "C"
    }
]

def run_quiz(questions):
    score = 0
    for question in questions:
        print(question["prompt"])
        for option in question["options"]:
            print(option)
        answer = input("Enter your answer: ").strip().upper()
        if answer == question["answer"]:
            print("Correct\n")
            score += 1
        else:
            print("Incorrect")

        print()
        
    print(f"Your score is {score} out of {len(questions)} questions correct.")

    print()
    
    if score >= 8:
        print("Scoring: Excellent! You have a great understanding of health.")
    elif score >= 5:
        print("Scoring: Good job! You have a solid knowledge of health.")
    elif score >= 3:
        print("Scoring: Fair. You might want to learn more about health topics.")
    else:
        print("Scoring: Needs Improvement. Consider studying more about health.")
        
run_quiz(questions)


```
